**Data Visualization**

**Bar Plot between Sub Category vs Category**

```python
plt.figure(figsize=(19,7))
plt.bar('Sub-Category','Category',data=df)
plt.show()
```

In furniture category the frequency of individual item is very less

In office category the frequency of individual item is medium

In Technology the frequency of individual item is high


**Pie Plot of Sub Category**

```python
plt.figure(figsize=(12,12))
df['Sub-Category'].value_counts().plot.pie(autopct='%1.1f%%')
plt.show()
```

The Store has wide variety of office supplies especially in Binders and paper department

**Bar Plot of Sales vs Profit**

```python
df.groupby('Sub-Category')['Profit','Sales'].agg(['sum']).plot.bar()
plt.title("total profit and sales per Sub-Category")
plt.rcParams['figure.figsize'] =[17,10]

plt.show()
```

Highest profit is earned in copiers while selling price of Chairs and Phones is extremely high as compared to other products.

People don't prefer to buy Tables and bookcases from Superstore

**Count Plot of States**

```python
plt.figure(figsize=(18,10))
sns.countplot(x=df['State'])
plt.xticks(rotation=90)
plt.title("State wise sale")
plt.show()
```

**Sales and profit region wise**

```python
df.groupby('Region')['Profit','Sales'].agg(['sum']).plot.bar()
plt.rcParams['figure.figsize']=[12,6]
plt.show()
```

**Count Plot of Sub-Categories**

```python
plt.figure(figsize=[12,6])
sns.countplot(x=df['Sub-Category'])
plt.xticks(rotation=90)
plt.show()
```

**Count Plot of Segment**

```python
sns.countplot(x=df['Segment'])
```

Consumer has more interest in order to buy more products as compare to other two segment.

Hence business should focus more on goods quality when it comes to sale it to the consumer segment.

in order to maintain and increase the sale with consumer segment.


**Count Plot of Region**

```python
sns.countplot(x=df['Region'])
```


'West' and 'East' regions are inclining towards buying more products as compare to 'South' and 'Central'.

Company should focuses more on it's quality in these regions and make stratergy to maintain the profit.


**Line Plot of Discount Vs Profit**

```python
plt.figure(figsize=(10,4))
sns.lineplot('Discount','Profit', data=df , color='b',label='Discount');
plt.legend();
plt.show();
```

**Pair Plot of Sub-Category**

```python
figsize=(15,10)
sns.pairplot(df,hue='Sub-Category');
plt.show
```

**Box Plot of Sales,Quality,Discount & Profit**

```python
df.plot(kind='box',subplots=True,layout=(3,2),sharex=False,sharey=False)
plt.rcParams['figure.figsize']=[14,10]
plt.show()
```
